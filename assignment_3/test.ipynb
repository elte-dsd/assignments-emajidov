```python
from DynamicWeightedMajority import DynamicWeightedMajority
from Stagger import Stagger
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
import numpy as np
```

```python
stagger = Stagger(120) 
x_train, y_train = stagger.generateData() 
x_test, y_test = stagger.generateData() 
```

```python
dwm =  DynamicWeightedMajority(0.5,0.01,6)
```

```python
size, accs, nb_accs = dwm.fit(x_train, y_train)
```

```python
plt.figure(1, figsize=(16,10))
plt.title("Acurracies")
plt.plot(np.arange(len(accs)), accs, label="DWM")
plt.legend()

plt.title("Acurracies")
plt.plot(np.arange(len(nb_accs)), nb_accs, label="Simpe NB")
plt.legend()
plt.show()
```

```python
plt.figure(1, figsize=(16,10))
plt.title("Experts")
plt.plot(np.arange(len(size)), size)
plt.legend()
plt.show()
```

```python
nb = GaussianNB() 
```

```python
nb.fit(x_train, y_train)
```

```python
preds = nb.predict(x_test)
nb_acc = accuracy_score(y_test, preds)
print(nb_acc)
```

```python
dwm_preds = dwm.predict(x_test)
dwm_acc = accuracy_score(y_test, dwm_preds)
print(dwm_acc)
```

```python
from skmultiflow.data.sea_generator import SEAGenerator
```

```python
stream = SEAGenerator(classification_function = , random_state = 112, balance_classes = False, noise_percentage = 0.28)
```

```python
stream.prepare_for_use()
```

```python
dwm = DynamicWeightedMajority(0.5,0.01,10)
dwm_accs = []
nb_accs = [] 
sizes = [] 
while(stream.has_more_samples()):
    data = stream.next_sample(10)
    size, dwm_acc, nb_acc = dwm.fit(data[0], data[1])
    for i in size: 
        sizes.append(i) 
    for i in dwm_acc:
        dwm_accs.append(i)
    for i in nb_acc: 
        nb_accs.append(i)
    
    
```

```python
plt.figure(1, figsize=(16,10))
plt.title("Acurracies")
plt.plot(np.arange(len(dwm_accs)), dwm_accs, label="DWM")
plt.legend()

plt.title("Acurracies")
plt.plot(np.arange(len(nb_accs)), nb_accs, label="Simpe NB")
plt.legend()
plt.show()
```
